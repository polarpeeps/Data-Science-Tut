### JSON TO CSV

```python
import pandas as pd
import json

sample_json = '{"users": [{"id": 1,"name": "Nabeel Rizvi","email": "nabeel@example.com","isActive": true,"roles": ["Admin", "Developer"]},{"id": 2,"name": "Sara Khan","email": "sara@example.com","isActive": false,"roles": ["Tester"]}],"meta": {"count": 2,"source": "Internal Database"}}'

# Load JSON string into a Python dict
data = json.loads(sample_json)

# Extract only the list under "users"
df = pd.json_normalize(data["users"])
df
```
