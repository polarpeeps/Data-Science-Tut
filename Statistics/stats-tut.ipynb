```python
## Z Score
outliers=[]

def detect_outliers(data):
  threshold=3 ## 3 std deviation
  mean=np.mean(data)
  std=np.std(data)

  for i in data:
    z_score=(i-mean)/std
    if np.abs(z_score)> threshold:
      outliers.append(i)

  return outliers

```

```python
detect_outliers(dataset)
```
