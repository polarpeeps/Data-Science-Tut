```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Graph in 2-D')
plt.show()
```

```python
y= 2*x*x
plt.plot(x,y,'b--')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'ro--')
plt.subplot(2,2,2)
plt.plot(x,y,'g^--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
plt.show()
```

```python
x=np.arange(1,11)
y  = 3*x +5
plt.plot(x,y,'b')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Test")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y  = np.sin(x)
plt.plot(x,y,'b')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("SIN-WAVE FORM")
plt.show()
```

### BAR PLOT

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### HISTOGRAMS

```python
a = np.array([20,25,46,23,42,22,24,26,16])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

### BOX PLOT USING MATPLOTLIB

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```
