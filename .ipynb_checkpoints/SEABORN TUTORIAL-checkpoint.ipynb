### Correlation with heatmap
**A correlation heatmap uses colored cells, typically in a monochromatice scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very imp in feature selection.**
- The method df.corr() works only on numeric columns, so when it encounters strings like 'No', 'Sun', 'Dinner', etc., it throws the error:

````python
df.corr(numeric_only=True)
````

````python
sns.heatmap(df.corr(numeric_only=True))
````

### DISTPLOT
**Definition:**  
`distplot()`  or `displot()`was used to visualize the distribution of a single variable using a histogram and a KDE (Kernel Density Estimate) curve.


**Definition:**  
`distplot()` was used to visualize the distribution of a single variable using a histogram and a KDE (Kernel Density Estimate) curve.

**Example:**
```python


````python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.distplot(df["total_bill"], bins=20, kde=True)
plt.show()
````



### JOINPLOT
**Definition:**  
`jointplot()` visualizes the relationship between two variables along with their individual distributions.

**Example:**


````python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.jointplot(x="tip", y="total_bill", data=df, kind="hex")
plt.show()
````

````python
### regression line
````

````python
sns.jointplot(x="tip", y="total_bill", data=df, kind="reg")
plt.show()
````


### PAIRPLOT
**Definition:**  
`pairplot()` creates pairwise relationships among numerical columns in a dataset.

**Example:**



````python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.pairplot(df, hue="smoker")
````

### Categorical Plots

````python
# COUNT PLOT
sns.countplot(x='sex',data=df)
````

````python
sns.countplot(x='day',data=df)
````

````python
# BAR PLOT
sns.barplot(y='total_bill',x='sex',data=df)
````

### BOX PLOT

````python
sns.boxplot(x='sex',y='total_bill', data= df, palette='rainbow')
````

````python
sns.boxplot(data=df,orient='v')
````

````python
sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
````

### VIOLIN PLOT

````python
# helps us to see both the distribuition of data in terms of kernel density estimation and boxplot
sns.violinplot(x="total_bill",y='day',hue='sex',data=df,palette='rainbow')
````
